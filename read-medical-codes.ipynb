```python
import numpy as np
import pandas as pd

cpt_codes = pd.read_excel("data/cpt-codes-jan-2023.xlsx")
```

# CPT Codes

```python
cpt_codes.shape
```

```python
cpt_list = []
for row in range(0,1114):
    if len(str(cpt_codes["Unnamed: 0"][row]).strip()) in range(5,7):
        cpt_list.append(str(cpt_codes["Unnamed: 0"][row]).strip())
        print(cpt_codes["Unnamed: 0"][row])
        print(cpt_codes["Unnamed: 1"][row])
    else:
        print("not a code row")
```

```python
cpt_codes = cpt_codes.astype(str)
```

```python
print(len(cpt_list))
```
